```python
from __future__ import print_function, division
from PIL import ImageFont, ImageDraw, Image
from random import randint, choice, uniform
from datetime import datetime as UTC
from std_msgs.msg import String
import numpy as np
import argparse
import json
import sys
import math
import cv2
import csv
import os
import re
import subprocess
import tempfile
from reportlab.graphics import renderPDF
from reportlab.graphics.barcode import code39, code128
from reportlab.pdfgen import canvas
from reportlab.lib.units import mm, cm, inch
import reportlab.rl_config
reportlab.rl_config.warnOnMissingFontGlyphs = 0
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
import qrcode
from PyPDF2 import PdfFileMerger

def create_labels():
    number_add_pdf = 40
    dpi=600
    max_pages = str(2*number_add_pdf)
    pdf_list=list()
    pg=0
    for lbl in range(number_add_pdf):
        lbl_img = Image.open('/tmp/AprilTag_Bundle_2x_tag36h11/tag_bundle_{:d}.png'.format(lbl)).convert('L')
        wlblimg, hlblimg = lbl_img.size
        
        lbl_img_upper = lbl_img.crop((0,0,wlblimg,int(hlblimg/2.0)))
        save_str='/tmp/LBL_'+str(pg)+'.pdf'
        lbl_img_upper.save(save_str, "PDF", resolution=600)
        print('Page: '+str(pg+1)+'/'+max_pages)
        pdf_list.append(save_str)
        pg+=1
        
        lbl_img_lower = lbl_img.crop((0,int(hlblimg/2.0),wlblimg,hlblimg))
        save_str='/tmp/LBL_'+str(pg)+'.pdf'
        lbl_img_lower.save(save_str, "PDF", resolution=600)
        print('Page: '+str(pg+1)+'/'+max_pages)
        pdf_list.append(save_str)
        pg+=1


    print('Merging PDF files')
    merger = PdfFileMerger()
    for pdf in pdf_list:
        merger.append(pdf)
    merger.write("/tmp/LBS.pdf")
    print('Finished merging')
    return(True)
create_labels()
```
